```python
import re
import numpy as np
import pandas as pd
```

```python
file_path = 'codebook_97.txt'
with open(file_path) as file:
    text = file.read()
```

```python
clean = []

for item in text.split('\n'):
    new = re.sub('\.', '', item)
    new = re.sub('\.', '', new)
    new = re.sub('      ', '  ', new)
    new = re.sub('     ', '  ', new)
    new = re.sub('     ', '  ', new)
    new = re.sub('    ', '  ', new)
    new = re.sub('   ', '  ', new)
    new = re.sub('   ', '  ', new)
    new = re.sub('   ', '  ', new)
    new = re.sub('   ', '  ', new)
    new_list = new.split('  ')
    out = {}
    try:
        out['ref'] = new_list[0]
        out['year'] = new_list[1]
        out['var'] = new_list[2]
        out['question'] = new_list[3]
    except:
        out['ref'] = np.nan
        out['year'] = np.nan
        out['var'] = np.nan
        out['question'] = np.nan
    clean.append(out)
```

```python
codes = pd.DataFrame(clean)
codes_have_year = codes[(codes.year != 'PSTR') & (codes.year != 'XRND') & (codes.year.notnull())]
codes_no_year = codes[(codes.year == 'PSTR') | (codes.year == 'XRND') | (pd.isnull(codes.year) == True)]

codes_no_year['test'] = codes_no_year['year'].str.extract(r'(\d{4})\s', expand=False)

# codes_no_year[codes_no_year.test.notnull()]
codes
```
